A point estimate is a single value estimate of a unknown population parameter. For instance, a sample mean is a point estimate of a population mean. An interval estimate gives you a range of values where the parameter is expected to lie. A confidence interval is the most common type of interval estimate.

# question 2


```python
def estimate_population_mean(sample_mean, sample_std_dev, sample_size):
    # Calculate the standard error of the mean
    std_error_mean = sample_std_dev / (sample_size ** 0.5)
    
    # Calculate the lower and upper bounds of the 95% confidence interval
    lower_bound = sample_mean - 1.96 * std_error_mean
    upper_bound = sample_mean + 1.96 * std_error_mean
    
    # Return the estimated population mean
    return round(sample_mean, 2), round(lower_bound, 2), round(upper_bound, 2)

estimate_population_mean(169.5 , 2.31 , 36)
```

Hypothesis testing is used to assess the plausibility of a hypothesis by using sample data. The test provides evidence concerning the plausibility of the hypothesis, given the data. Statistical analysts test a hypothesis by measuring and examining a random sample of the population being analyzed.

Hypothesis testing allows the researcher to determine whether the data from the sample is statistically significant. Hypothesis testing is one of the most important processes for measuring the validity and reliability of outcomes in any systematic investigation

# question 5


```python
import numpy as np
from scipy.stats import ttest_ind

# Generate two samples
male_weight = np.array([68, 72, 76, 80, 84]) # sample of male college students
female_weight = np.array([52, 56, 60, 64, 68]) # sample of female college students

# Calculate sample statistics
male_mean = np.mean(male_weight)
female_mean = np.mean(female_weight)
male_std = np.std(male_weight, ddof=1)
female_std = np.std(female_weight, ddof=1)
male_n = len(male_weight)
female_n = len(female_weight)

# Conduct two-sample t-test
t, p = ttest_ind(male_weight, female_weight, equal_var=False)

# Print results
print("Male mean weight:", male_mean)
print("Female mean weight:", female_mean)
print("Male standard deviation:", male_std)
print("Female standard deviation:", female_std)
print("Male sample size:", male_n)
print("Female sample size:", female_n)
print("t-statistic:", t)
print("p-value:", p)

alpha = 0.05
if p < alpha:
    print("We reject the null hypothesis.")
else:
    print("We fail to reject the null hypothesis.")

```

If the p value is less than significance value then the null hypothesis is rejected. Otherwise, it is accepted.

```python
import matplotlib.pyplot as plt
from scipy.stats import t
import numpy as np

# set the degrees of freedom
df = 10

# generate x values for the plot
x = np.linspace(t.ppf(0.001, df), t.ppf(0.999, df), 1000)

# generate the y values for the plot (the probability density function)
y = t.pdf(x, df)

# plot the t-distribution
plt.plot(x, y)

# set the plot title and axis labels
plt.title("Student's t-Distribution (df=10)")
plt.xlabel("x")
plt.ylabel("PDF")

# display the plot
plt.show()
```

```python
# question 10

import numpy as np
from scipy.stats import ttest_ind

# generate two random samples of equal size
sample1 = np.random.normal(loc=5, scale=2, size=100)
sample2 = np.random.normal(loc=5, scale=2, size=100)

# calculate the t-statistic and p-value using scipy's ttest_ind function
t_statistic, p_value = ttest_ind(sample1, sample2)

# print the results
print("T-Statistic: ", t_statistic)
print("P-Value: ", p_value)

# check for statistical significance at the alpha=0.05 level
if p_value < 0.05:
    print("Reject the null hypothesis - there is evidence of a significant difference in the population means.")
else:
    print("Fail to reject the null hypothesis - there is no evidence of a significant difference in the population means.")
```

# question 11

Student's t-distribution is a probability distribution that arises when estimating the mean of a normally distributed population in situations where the sample size is small, or when the population variance is unknown. It is named after William Gosset, who wrote under the pseudonym "Student".

The t-distribution looks similar to a normal distribution, but with heavier tails. The shape of the t-distribution depends on the sample size, with larger sample sizes leading to distributions that are increasingly similar to the normal distribution.

The t-distribution is commonly used in hypothesis testing and confidence interval estimation. It is used when the sample size is small and/or the population variance is unknown. Specifically, when the sample size is less than 30 or the population standard deviation is unknown, the t-distribution should be used instead of the standard normal distribution. The t-distribution is also used in linear regression analysis, where it is used to test hypotheses about the regression coefficients.
